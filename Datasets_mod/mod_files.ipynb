```python
# Convirtiendo json a csv usando Pandas

import pandas as pd
import json

with open('../Datasets_mod/constructors.json',encoding='UTF-8') as reader:
    data = json.load(reader)
constructors_df = pd.DataFrame(data)
constructors_df.head()
```

```python
with open('../Datasets_mod/pit_stops.json',encoding='UTF-8') as reader:
    data = json.load(reader)
pitstops_df = pd.DataFrame(data)
pitstops_df.head()
```

```python
with open('../Datasets_mod/results.json',encoding='UTF-8') as reader:
    data = json.load(reader)
results_df = pd.DataFrame(data)
results_df.head()
```

```python
#with open('../Datasets_mod/drivers.json',encoding='UTF-8') as reader:
#    data = json.load(reader)
#drivers_df = pd.DataFrame(data)
#drivers_df.head()

drivers_df = pd.read_json('../Datasets_mod/drivers.json',lines=True, encoding='UTF-8')
drivers_df = pd.concat([drivers_df.drop(['name'], axis=1), drivers_df['name'].apply(pd.Series)], axis=1)
#new_cols = ["driverId",    "driverRef"    ,"number",    "code","forename",    "surname",    "dob",     "nationality",    "url"]
#drivers_df = drivers_df.reindex(columns=new_cols)
drivers_df.head(2)
```

```python
#https://stackoverflow.com/questions/35711059/extract-dictionary-value-from-column-in-data-frame

drivers_df = pd.concat([drivers_df, pd.DataFrame(list(drivers_df['name']))], axis=1)
drivers_df.head()
```

```python

drivers_df.drop(['name'], axis=1, inplace=True)
drivers_df.head()
```

```python
with open('../Datasets_mod/Qualifying/qualifying_split_1.json',encoding='UTF-8') as reader:
    data = json.load(reader)
Q1_df = pd.DataFrame(data)
Q1_df.head()
```

```python
with open('../Datasets_mod/Qualifying/qualifying_split_2.json',encoding='UTF-8') as reader:
    data = json.load(reader)
Q2_df = pd.DataFrame(data)
Q2_df.tail()
```

```python
# Consolidando archivo de lap_times

data1 = pd.read_csv('../Datasets_mod/CSV/lap_times/lap_times_split_1.csv')
lts1_df = pd.DataFrame(data1)

data2 = pd.read_csv('../Datasets_mod/CSV/lap_times/lap_times_split_2.csv')
lts2_df = pd.DataFrame(data2)

data3 = pd.read_csv('../Datasets_mod/CSV/lap_times/lap_times_split_3.csv')
lts3_df = pd.DataFrame(data3)

data4 = pd.read_csv('../Datasets_mod/CSV/lap_times/lap_times_split_4.csv')
lts4_df = pd.DataFrame(data4)

data5 = pd.read_csv('../Datasets_mod/CSV/lap_times/lap_times_split_5.csv')
lts5_df = pd.DataFrame(data5)


lap_times_df=pd.concat([lts1_df,lts2_df, lts3_df, lts4_df, lts5_df])
```
